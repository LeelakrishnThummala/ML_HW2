```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/HW2_linear_data.csv')
```

```python
data = pd.read_csv('/content/HW2_linear_data.csv')
X = data['X'].values
Y = data['Y'].values
```

```python
data
```

```python
# Initialize parameters (m and c)
m = 0
c = 0

# Hyperparameters
learning_rate = 0.0001
epochs = 1000
```

```python
def mse_loss(Y, Y_pred):
    return ((Y - Y_pred) ** 2).mean()

# Lists to store loss values for plotting
loss_history = []

# Training the model using gradient descent
for epoch in range(epochs):
    # Calculate predictions
    Y_pred = m * X + c

    # Calculate the gradient of the loss with respect to m and c
    dm = (-2 / len(X)) * np.sum(X * (Y - Y_pred))
    dc = (-2 / len(X)) * np.sum(Y - Y_pred)

    # Update parameters using gradient descent
    m -= learning_rate * dm
    c -= learning_rate * dc

    # Calculate and store the current loss
    loss = mse_loss(Y, Y_pred)
    loss_history.append(loss)

# Plot the data points and the linear regression line
plt.scatter(X, Y, label='Data Points')
plt.plot(X, m * X + c, color='red', label='Linear Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression using Gradient Descent')
plt.legend()

# Show the plot
plt.show()

# Print the final values of m and c
print(f'Final value of m: {m}')
print(f'Final value of c: {c}')

# Print the final MSE loss
final_loss = loss_history[-1]
print(f'Final MSE loss: {final_loss}')
```
